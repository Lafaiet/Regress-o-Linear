O **Coeficiciente de correlação amostral de Pearson** $r$ pode ser obtido pela fórmula:

\begin{equation*}
r = \frac{\sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y}) } {\sqrt{\sum_{i=1}^n(x_i-\bar{x})^{2} \sum_{i=1}^n(y_i-\bar{y})^{2}}}
\end{equation*}

```python
# Imports
from __future__ import unicode_literals

from math import sqrt

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_correlation(x, y):
    x_mean = sum(x)/len(x)
    y_mean = sum(y)/len(y)
    
    A = sum(map( lambda x_y : ((x_y[0]-x_mean))*((x_y[1]-y_mean)), zip(x, y) ))
    B = sqrt(sum(map( lambda xi : (xi-x_mean)**2, x)) * sum(map( lambda yi : (yi-y_mean)**2, y)) )
    
    return float(A)/B
    
```

### Coeficientes de correlação das variáveis em relação ao preço:

```python
y = database['price']
```

## Tamanho:

```python
x = database['size']
r = get_correlation(x, y)
print r
```

```python
fig, ax = plt.subplots()
plt.xlabel('Tamanho')
plt.ylabel('Preço (R$)')
plt.title("Tamanho X Preço")
plt.figtext(1, .8, "r = %.2f"%r, size= 'x-large')

ax.scatter(x, y)
plt.plot()
```
